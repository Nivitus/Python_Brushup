### Lambda within Filter()

```python
my_list = [2,4,6,8,10]
newlist = list(filter(lambda a:(a/3==2), my_list))  # Syntax: filter(function,interable)
print(newlist)
```

### Lambda within Map()

```python
my_list = [2,4,6,8,10]
newlist = list(map(lambda a:(a/3==2), my_list))   # Syntax: 
print(newlist)
```
