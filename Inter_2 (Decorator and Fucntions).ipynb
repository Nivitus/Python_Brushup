# Functions in Python

```python
# Functions passing arguements:
def func1(name):
    return f"Hello{name}"
def func2(name):
    return f"{name}, how u doin?"
def func3(func4):
    return func4("hola Nivitus")

print(func3(func1))
print(func3(func2))
```

```python
# Inner Or Nested Functoins:
def func():
    print("Parent Function")
    
    def child_1():
        print("Fisrt Baby Function")
    def child_2():
        print("Second Baby Function")
    def child_3():
        print("Third Baby Function")
    
    child_1()
    child_2()
    child_2()
    
func()    
        
```

```python
def func(n):
    def func1():
        return "Nivitus"
    def func2():
        return "Isbella Palacios"
    if n == 1:
        return func1
    else:
        return func2
    
a = func(1)
b = func(2)
    
print(a())
print(b())
        
```

# Decorators in Python

```python
def function1(function):
    def wrappers():
        print("Hola")
        function()
        print("Welcome to Become a Data Scientist")
    return wrappers
def function2():
    print("Data Scientist")
    
function2 = function1(function2)
function2()
```

```python
# So this is called Decorators , as well as sentatic sugar:
def function1(function):
    def wrappers():
        print("Hola")
        function()
        print("Welcome to Become a Data Scientist")
    return wrappers
@function1
def function2():
    print("Data Scientist")
    
function2()    
```
