# While Loop

```python
count = 0
while count < 10:
    print("Count is", count)
    count = count + 1
print("Good bye")    
```

```python
import random
```

```python
n = 20
to_b_guessed = int(n * random.random()) + 1
guess = 0
while guess != to_b_guessed:
    guess = int(input("New Number: "))
    if guess < 0:
        if guess > to_b_guessed:
            print("Number too Large")
        elif guess < to_b_guessed:
            print("Number too Small")
    else:
        print("Sorry! that you're giving up")
        break
else:
    print("Congrats! You made it!")        
```

# For Loop

```python
cars = ["Audi","BMW","Benz","Ferrari"]
for car in cars:
    print("I love this car: ", car)
print("Have a good day!")    
```

```python
num = int(input("Enter the Number:"))
fact = 1
if num < 0:
    print("NUmbers should be possitive only!")
elif num == 0:
    print(" Fact is = 1")
else:
    for i in range(1, num + 1):
        fact = fact * i
print(fact)        
    
```

# Nested Loop

```python
travelling  = input("yes or no:")
while travelling == "yes":
    num = int(input("number of the people travelling:"))
    
    for i in range(1, num + 1):
        name = input("Enter the Name")
        Age = input("Enter the Age")
        Sex = input("Enter the Sex")
        
        print(name)
        print(Age)
        print(Sex)
    travelling = input("Oops! forgot anyone for travel")        

```
