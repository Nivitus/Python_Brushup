```python
from collections import namedtuple
```

```python
# Here is a namedtuple model
a = namedtuple('zip','name, tech')
s = a('data science','python')
print(s)
```

```python
a = namedtuple('zip','name, tech')
s = a._make(['data science','python'])
print(s)
```

```python
from collections import ChainMap
```

```python
# ChainMap
a = {1:'AI',2:'DL',3:'ML'}
b = {4:'DS',5:'CV',6:'NLP'}
all = ChainMap(a,b)
print(all)
```

```python
from collections import Counter
```

```python
# Counter
AI = [2,2,5,6,4,2,6,8,1,4,3,1,7,9,2,2,0]
count = Counter(AI)
print(count)
```

```python
# Types of Counter Operations
print(list(count.elements()))
```

```python
print(count.most_common())
```

```python
sub = {2:2,4:1} 
```

```python
print(count.subtract(sub))
```

```python
print(count.most_common())
```
